Функция apply() и ее применение

```python
import pandas as pd

data = pd.read_csv('data/data.csv')
melb_df = data.copy()

#приведем к единому временному формату
melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst=True)
#создадим новый признак
melb_df['WeekdaySale'] = melb_df['Date'].dt.weekday

weekend_count = melb_df['WeekdaySale'].value_counts(normalize=False)
#display(weekend_count)


#задание 1
#Напишите функцию get_weekend(weekday), которая принимает на вход элемент столбца WeekdaySale 
# и возвращает 1, если день является выходным, и 0 — в противном случае, 
# и создайте столбец Weekend в таблице melb_df с помощью неё. Далее найти среднюю подажу в выходной день
def get_weekend(weekday):
    if weekday == 5 or weekday == 6:
        return 1
    else:
        return 0
    
melb_df['Weekend'] = melb_df['WeekdaySale'].apply(get_weekend)
mean_price = melb_df[melb_df['Weekend'] == 1]['Price'].mean()
print('Средняя цена обьекта проданного в выходной день:')
print(round(mean_price)) 

#задание 2
#Преобразуйте столбец SellerG с наименованиями риелторских компаний в таблице melb_df следующим образом: 
# оставьте в столбце только 49 самых популярных компаний, а остальные обозначьте как 'other'

#получим список-индекс из топа 49 контор
top_sellers = melb_df['SellerG'].value_counts().nlargest(49).index
#функция проверки принадлежности к топ конторам
def get_top_seller (seller):
    if seller in top_sellers:
        return seller
    else: 
        return 'Other'
#теперь мы чутка модернизируем столбец SellerG
melb_df['SellerG'] = melb_df ['SellerG'].apply(get_top_seller)
#посмотрим на уникальные значения, всего должно получится 49 значений
melb_df['SellerG'].nunique()

#задание 2_1
#Найдите, во сколько раз минимальная цена объектов недвижимости, проданных компанией 'Nelson', 
# больше минимальной цены объектов, проданных компаниями, обозначенными как 'other'. Ответ округлите до десятых
sum_Nelson = melb_df[melb_df['SellerG'] == 'Nelson']['Price'].min()
sum_Other = melb_df[melb_df['SellerG'] == 'Other']['Price'].min()
print(f'Минимальная цена обьета компании Тельсон равна {sum_Nelson}, в то время, как минимальная цена других компаний равна {sum_Other}, что в {round(sum_Nelson/sum_Other, 1)} меньше')
 
#задание 3
#написать функцию извлечения из Series кол месяцев

test_series_2 = pd.Series([
    'Опыт работы 5 лет',        #60
    'Опыт работы 5 месяцев',    #5
    'Опыт работы 1 год 1 месяц',#13
    'Опыт работы 3 месяца',     #3
    'Опыт работы 7 лет'         #84
])

def get_experience(arg):
  
    #задаем списки поиска месяцев и годов
    lst_mon = ['месяцев', 'месяца', 'месяц']
    lst_yer = ['год', 'лет','годов']
    #основной цикл по структуре даных
    
    #задаем переменную с месяцами
    month = 0
    #строчку бьем в список слов
    list_word = arg.split()
    #Побежали по списку, берем каждое слов и смотри не месяц это или год
    for word in list_word:
        if word in lst_yer:
            #если это год,значит перед ним цифра в годах
            value_index = list_word.index(word)
            month = int(list_word[value_index-1]) * 12
        if word in lst_mon:
        #если это месяц значит перед ним 
            value_index = list_word.index(word)
            month = month + int(list_word[value_index-1])
               
    return month 


print(test_series_2.apply(get_experience))
```
