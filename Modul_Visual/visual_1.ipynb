```python
import pandas as pd


covid_data = pd.read_csv ('Data/covid_df.csv')
# линейный график
grouped_cases = covid_data.groupby('date')['daily_confirmed'].sum()
grouped_cases.plot(
    kind='line',
    figsize=(12, 4),
    title='Ежедневная заболеваемость во времени',
    grid = True,
    lw = 3)     # толщина линии на графике

# столбчатая диаграмма 
# группируем по странам, берем столбик заболевших, вовыди послдение значения в столбце 
grouped_country = covid_data.groupby(['country'])['confirmed'].last()
grouped_country = grouped_country.nlargest(15)
grouped_country.plot(
    kind='bar',
    grid=True,
    figsize=(12, 4),
    colormap='plasma'   # цветовая гамма графика
)

```
